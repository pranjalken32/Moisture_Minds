Model 1 for df1

```python
Y=df1['sm']
X=df1.drop(columns=['sm','ttime'],axis=1)
```

```python
X.head()
```

```python
Y.head()
```

```python

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=16)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
dtr = DecisionTreeRegressor(max_depth=5)
rfr1 = RandomForestRegressor(n_estimators=100, random_state=16)
svr = SVR(kernel='sigmoid', gamma=1.0)
```

```python
lr2 = LinearRegression()
dtr2 = DecisionTreeRegressor(max_depth=5)
rfr2 = RandomForestRegressor(n_estimators=100, random_state=16)
svr2 = SVR(kernel='sigmoid', gamma=1.0)
```

```python
clfs1 = {
    'LR' : lr,
    'DTR' : dtr, 
    'RFR': rfr1,
    'SVR' : svr
}
```

```python
clfs2 = {
    'LR' : lr,
    'DTR' : dtr, 
    'RFR': rfr2,
    'SVR' : svr
}
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

```

```python
def train_classifier(clf,X_train,y_train,X_test,y_test):
    clf.fit(X_train,y_train)
    Y_pred = clf.predict(X_test)
    mae=  mean_absolute_error(Y_test,Y_pred)
    mse = mean_squared_error(Y_test,Y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(Y_test,Y_pred)
    
    return mae,mse,rmse,r2
```

```python
mae = []
mse = []
rmse = []
r2 = []

for name,clf in clfs1.items():
    
    curr_mae,curr_mse,curr_rmse,curr_r2 = train_classifier(clf, X_train,Y_train,X_test,Y_test)
    
    print("For ",name)
    print("Mean Absolute Error - ", curr_mae)
    print("Mean Square Error - ", curr_mse)
    print("Root Mean Square Error - ",curr_rmse)
    print("R2 Score - ", curr_r2)
    print(end="\n")
    
    mae.append(curr_mae)
    mse.append(curr_mse)
    rmse.append(curr_rmse)
    r2.append(curr_r2)
```

```python
model1 = RandomForestRegressor(n_estimators=100, random_state=16)
model1.fit(X_train, Y_train)
```

```python
Y_pred = model1.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
print('Root Mean Squared Error:', rmse)
```
